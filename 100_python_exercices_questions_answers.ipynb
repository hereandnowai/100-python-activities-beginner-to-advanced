````python
import math
def square_root(n):
  return math.sqrt(n)
print(square_root(16))

# Babylonian method (also known as Heron's method)
def manual_square_root(n, iterations=1000):
    guess = n / 2.0
    for i in range(iterations):
        guess = (guess + n / guess) / 2.0
    return guess

print(manual_square_root(16))
````

6. Write a Python program to solve quadratic equations.

````python
import cmath
def solve_quadratic(a, b, c):
  d = (b**2) - (4*a*c)
  sol1 = (-b - cmath.sqrt(d)) / (2*a)
  sol2 = (-b + cmath.sqrt(d)) / (2*a)
  return sol1, sol2
print(solve_quadratic(1, -7, 12))
````

````python
def is_prime(n):
  if n <= 1:
    return False
  for i in range(2, n): # for i in range(2, int(n**0.5) + 1):
    if n % i  == 0:
      return False
  return True
print(is_prime(16))

def is_prime_sqrt_method(n):
  if n<= 1:
    return False
  for i in range(2, int(n**0.5)+1):
    if n % i == 0:
      return False
    return True
print(is_prime(49))
````

15. Write a Python program to print all prime numbers in an interval.

````python
def print_primes(start, end):
  primes = []
  for num in range(start, end + 1):
    if is_prime(num):
      primes.append(num)
  return primes
print(print_primes(10, 50))
````

````python
# Method 1
import math
print(math.factorial(5))

# Method 2
def factorial(n):
  if n == 0:
    return 1
  else:
    return n * factorial(n-1)
print(factorial(5))

# Method 3
def fac_m3(n):
  result = 1
  for i in range(1, n+1):
    result *= i
  return result
print(fac_m3(5))
````

18. Write a Python program to print the Fibonacci sequence.

````python
n = 8
max_value = n

def fibonacci(n):
  fib_seq = [0, 1]
  while len(fib_seq) < n:
    fib_seq.append(fib_seq[-1] + fib_seq[-2])
  return fib_seq
print(fibonacci(n))

def fibonacci_up_to_n(max_value):
  fib_seq = [0, 1]
  while True:
    next_value = fib_seq[-1] + fib_seq[-2]
    if next_value > max_value:
      break
    fib_seq.append(next_value)
  return fib_seq
print(fibonacci_up_to_n(max_value))
````

19. Write a Python program to check Armstrong number.

````python
def is_armstrong(n):
  num_str = str(n)
  num_len = len(num_str)
  total = sum(int(digit)**num_len for digit in num_str)
  return n == total
print(is_armstrong(153))
````

21. Write a Python program to find the factors of a number.

````python
# Method 1
def factors(n):
  return [i for i in range(1, n+1) if n % i == 0]
print(factors(16))

# Method 2
def factors_m2(n):
  result = set()
  for i in range(1, int(n**0.5) + 1):
    if n % i == 0:
      result.add(i)
      result.add(n // i)
  return sorted(result)

print(factors_m2(16))
````

````python
#Method 1
import math
def find_hcf(a, b, c):
  return math.gcd(a, b, c)
print(find_hcf(32, 48, 64))

#Method 2
def gcd(a, b):
  while b:
    a, b = b, a % b
  return a
print(gcd(54, 24))
````

29. Write a Python program to check if a number is palindrome or not.

````python
# for a number
def is_palindrome(num):
  num = str(num) # Convert the number to a string
  return num == num[::-1] # check if the string is equal to its reverse
print(is_palindrome("Madam"))

# for a word
import re
def is_palindrome(s):
  s = re.sub(r'[^a-zA-Z0-9]', "", s).lower() # convert the string to lowercase
  return s == s[::-1] # check if the string is equal to its reverse

print(is_palindrome("Race Car"))
print(is_palindrome("Hello"))
````

30. Write a Python program to print Pascal’s triangle.

````python
def pascal_triangle(n):
  result = []
  for i in range(n):
    row = [1] * (i + 1)
    for j in range(1, i):
      row[j] = result[i - 1][j - 1] + result[i -1][j]
    result.append(row)
  return result
print(pascal_triangle(6))
````
